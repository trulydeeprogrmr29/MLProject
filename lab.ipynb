DATA PROCESSING


```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# First, we have performed one hot encoding on the categorical features, and then
# performed standard scaling on all the features
prty=prty.drop('Year',axis=True)
X,y=prty.drop('hg/ha_yield',axis=1),prty['hg/ha_yield']
X = pd.get_dummies(X,columns = cat_cols, drop_first = True)
scaler = StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

DATA SPLITTING

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

RANDOM_STATE = 42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=RANDOM_STATE)

print(X_train.shape,X_test.shape)
```

XGBOOST BASED MODEL

```python
import numpy as np
import xgboost as xgb
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

BASELINE XGBOOST MODEL

```python
xgb_baseline = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_baseline.fit(X_train, y_train)
y_pred_baseline = xgb_baseline.predict(X_test)
```

BASELINE PERFORMANCE MATRIX


```python
r2_baseline = r2_score(y_test, y_pred_baseline)
mae_baseline = mean_absolute_error(y_test, y_pred_baseline)
mse_baseline = mean_squared_error(y_test, y_pred_baseline)
rmse_baseline = np.sqrt(mse_baseline)

print("Baseline XGBoost Model Performance:")
print(f"R²: {r2_baseline:.4f}")
print(f"MAE: {mae_baseline:.4f}")
print(f"MSE: {mse_baseline:.4f}")
print(f"RMSE: {rmse_baseline:.4f}")
```

HYPERPARAMETER tuning using GridSeARCHCV

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}

xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
```

```python
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)
```

```python
print("\nBest hyperparameters found:")
print(grid_search.best_params_)
```

Train with best parameter

```python
xgb_tuned = grid_search.best_estimator_
y_pred_tuned = xgb_tuned.predict(X_test)
```

TUned Performance Metrices

```python
r2_tuned = r2_score(y_test, y_pred_tuned)
mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
mse_tuned = mean_squared_error(y_test, y_pred_tuned)
rmse_tuned = np.sqrt(mse_tuned)

print("\nTuned XGBoost Model Performance:")
print(f"R²: {r2_tuned:.4f}")
print(f"MAE: {mae_tuned:.4f}")
print(f"MSE: {mse_tuned:.4f}")
print(f"RMSE: {rmse_tuned:.4f}")
```

Scatterplot for XGBoost Baseline Model Predictions

```python
plt.figure(figsize=(12, 5))

sns.set_style("white")
sns.scatterplot(x=y_test, y=y_pred_baseline, alpha=0.7, color='blue', label='Predicted')
sns.scatterplot(x=y_test, y=y_test, alpha=0.7, color='red', label='Actual')
plt.xlabel("Actual Yield")
plt.ylabel("Predicted Yield")
plt.title("XGBoost Baseline Model Predictions")
plt.grid(True)
plt.legend()
plt.savefig("baseline_model_predictions.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.set_style("white")

plt.figure(figsize=(12, 5), facecolor='white')
sns.scatterplot(x=y_test, y=y_pred_tuned, color='green', alpha=0.7, label='Predicted')
diagonal = np.linspace(y_test.min(), y_test.max(), 100)
sns.scatterplot(x=diagonal, y=diagonal, color='red', alpha=0.7, s=50, label='Actual')

plt.xlabel("Actual Yield")
plt.ylabel("Predicted Yield")
plt.title("XGBoost Tuned Model Predictions")
plt.grid(True)
plt.legend()
plt.savefig("xgb_tuned_model_predictions.png", dpi=300, bbox_inches='tight')
plt.show()
```

Residual plots for XGBoost Baseline and Tuned Models

```python
baseline_residuals = y_test - y_pred_baseline
tuned_residuals = y_test - y_pred_tuned

sns.set_style("white")

plt.figure(figsize=(12, 5), facecolor='white')
sns.scatterplot(x=y_pred_baseline, y=baseline_residuals, color='blue', alpha=0.7, label='Residuals')
x_range_baseline = np.linspace(y_pred_baseline.min(), y_pred_baseline.max(), 50)
sns.scatterplot(x=x_range_baseline, y=np.zeros_like(x_range_baseline), 
                color='red', marker='o', s=50, label='Zero Error')

plt.xlabel("Predicted Yield")
plt.ylabel("Residuals")
plt.title("Baseline XGBoost Model Residuals")
plt.savefig("xgb_baseline_residuals.png", dpi=300, bbox_inches='tight')
plt.legend()

plt.figure(figsize=(12, 5), facecolor='white')
sns.scatterplot(x=y_pred_tuned, y=tuned_residuals, color='green', alpha=0.7, label='Residuals')
x_range_tuned = np.linspace(y_pred_tuned.min(), y_pred_tuned.max(), 50)
sns.scatterplot(x=x_range_tuned, y=np.zeros_like(x_range_tuned), 
                color='red', marker='o', s=50, label='Zero Error')

plt.xlabel("Predicted Yield")
plt.ylabel("Residuals")
plt.title("Tuned XGBoost Model Residuals")
plt.legend()
plt.savefig("xgb_tuned_residuals.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
